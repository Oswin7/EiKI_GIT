````python
numberOfFunction = input('Gib entweder 1 oder 2 ein: ')
x1 = np.arange(1,20) # X-axis
y1 = np.arange(1,20)  # Y1-axis
y2 = 2**np.copy(y1)  # Y2-axis

if (numberOfFunction == '1'):
    plt.title("linear function")
    plt.xlabel("X-values")
    plt.ylabel("Y-values")
    plt.plot(x1, y1, marker='x', linestyle='', color='blue')
    plt.show()
elif (numberOfFunction == '2'):
    plt.title("exponential function")
    plt.xlabel("X-values")
    plt.ylabel("Y-values")
    plt.plot(x1, y2, marker='x', linestyle='', color='red')
    plt.show()
else:
    print('Wrong number!')
````
